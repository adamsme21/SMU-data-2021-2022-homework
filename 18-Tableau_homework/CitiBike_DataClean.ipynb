```python
import pandas as pd
import numpy as np
```

```python
months = ["01", "02", "03", "04", "05"]
df21 = pd.DataFrame()

for month in months:
    print(month)
    filepath = f"2021_data/2021{month}-citibike-tripdata.csv"
    df= pd.read_csv(filepath, low_memory=False)
    df = df.sample(frac= 0.015, random_state=42)
    df21 = df21.append(df)
    
df21 = df21.reset_index(drop=True)
```
